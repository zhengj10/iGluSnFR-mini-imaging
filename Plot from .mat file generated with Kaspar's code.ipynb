```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
matPath = '/Volumes/genie/Jihong_GENIE_stuff/iGluSnFR_spontaneous_activity/Kaspar code/summary/Kaspar code'
matFile = matPath+'/230222_Plate_001_633.9064_FOV_00001.mat'
```

```python
data = h5py.File(matFile,'r')
data.keys()
```

```python
dFF = data['A']['DFF'][()]
```

```python
x = np.arange(1,dFF.shape[0]+1)
y_shift = 0
plt.figure(figsize=[13,dFF.shape[1]/10.0])
for i in range(dFF.shape[1]):
    offset = y_shift
    if np.max(dFF[:,i])>10:
        continue
    plt.fill_between(x,dFF[:,i]-offset,0-offset,zorder=y_shift+1,facecolor='w',edgecolor='w',lw=0,closed=True)
    plt.plot(x,dFF[:,i]-offset,color='k',zorder=y_shift+1,lw=1)
    y_shift+=1
    plt.axis('off')
plt.tight_layout()
plt.show()


```
